```python
df_new = df.groupby(by=['title'])['rating'].count().reset_index()
```

```python
df_new.head()
```

```python
df_new.rename(columns={'rating': 'totalRatingCount'}, inplace=True)
df_new.head()
```

```python
df_2 = df.merge(df_new, left_on='title', right_on='title', how='left')
df_2.head()
```

```python
popularity_threshold_1 = 50
df_popular_1 = df_2.query('totalRatingCount >= @popularity_threshold_1')
df_popular_1.head()
```

```python
df_popular_1.shape
```

```python
popularity_threshold = 50
df_popular = df_2.query('totalRatingCount >= @popularity_threshold')
df_popular.head()
```

```python
df_popular.shape
```

```python
movie_features = df_popular.pivot_table(index='title', columns='userId', values='rating').fillna(0)
movie_features.head()
```

```python
movie_features_matrix = csr_matrix(movie_features.values)
movie_features_matrix
```

```python
# NearestNeighbors Algorithm
knn_model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=5, n_jobs=None)
knn_model.fit(movie_features_matrix)
```

```python
query_index = np.random.choice(movie_features.shape[0])
query_index
```

```python
distance, indices = knn_model.kneighbors(movie_features.iloc[query_index, :].values.reshape(1, -1), n_neighbors=10)
```

```python
len(distance.flatten())
```

```python
for i in range(0, len(distance.flatten())):
    if i == 0:
      print('Recommendations for {0}:\n'.format(movie_features.index[query_index]))
    else:
      print('{0} Recommendations for {1}: at distance of {2}\n'.format(i, movie_features.index[query_index], distance.flatten()[i]))
```
